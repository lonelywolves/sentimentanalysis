```python
def extract_info(responseJson):
  for k in responseJson['items']:

    try:
      ids.append(k['id']['videoId'])
    except:
      ids.append("err")
      titles.append("err")
      dates.append("err")
      continue

    try:
      titles.append(k['snippet']['title'])
    except:
      titles.append("err")
      dates.append("err")
      continue

    try:
      dates.append(k['snippet']['publishTime'])
    except:
      dates.append("err")
      continue
```

```python
get_response(url1)
get_response(url2)
get_response(url3)
get_response(url4)
get_response(url5)
get_response(url6)
get_response(url7)
get_response(url8)
get_response(url9)
get_response(url10)
get_response(url11)
```

```python
df = pd.DataFrame([ids,titles,dates]).T
df.columns = ['ids','titles','Date']
df = df.drop(df[(df['ids'] == 'err') | (df['titles'] == 'err') | (df['Date'] == "err")].index)

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M:%S')

df = df[(df['Date'].dt.year == 2019)|(df['Date'].dt.year == 2020)|(df['Date'].dt.year == 2021)]

df = df.reset_index(drop=True)
```

```python
for i in range(len(df['ids'])):
  idx = df['ids'][i].find("-")
  if idx==0:
    df.drop([i],axis=0,inplace = True)
df = df.reset_index(drop=True)
```
